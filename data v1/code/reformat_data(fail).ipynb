```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data 
acelerometro_terra =pd.read_csv("NEW_acelerometro_terra.csv")
campoMagnetico_terra =pd.read_csv("NEW_campoMagnetico_terra.csv")
giroscopio_terra = pd.read_csv("NEW_giroscopio_terra.csv")
aceleracaoLinear = pd.read_csv("NEW_aceleracaoLinear_terra.csv")
groundTruth = pd.read_csv("groundTruth.csv")

#aceleracaoLinear.head(30)
```

```python
#convert data to numpy array
np_all_acelerometro_terra = np.array(acelerometro_terra) 
np_all_campoMagnetico_terra = np.array(campoMagnetico_terra) 
np_all_giroscopio_terra = np.array(giroscopio_terra) 
np_all_aceleracaoLinear = np.array(aceleracaoLinear) 
np_groundTruth=np.array(groundTruth)
```

```python
# drop unimportant columns
acelerometro_terra = acelerometro_terra.drop(['Unnamed: 0','timestamp','val','uptimeNanos'] ,axis=1)
campoMagnetico_terra = campoMagnetico_terra.drop(['Unnamed: 0','timestamp','val','uptimeNanos'] ,axis=1)
giroscopio_terra = giroscopio_terra.drop(['Unnamed: 0','timestamp','val','uptimeNanos'] ,axis=1)
aceleracaoLinear = aceleracaoLinear.drop(['Unnamed: 0','timestamp','val','uptimeNanos'] ,axis=1)
```

```python
#convert data to numpy array
acelerometro_terra = np.array(acelerometro_terra)
campoMagnetico_terra = np.array(campoMagnetico_terra)
giroscopio_terra = np.array(giroscopio_terra)
aceleracaoLinear = np.array(aceleracaoLinear)
```

### Reformate data set that all the records that have the same y (output) to be on recorde 2D of input

- Reformate acelerometro data

```python
out_acelerometro_terra=[]
i=0
for elem in np_groundTruth :
    lap = elem[0]
    start = elem[1]
    end = elem[2]
    y2 =[np.array([])]
    y1 = []
    #get the noemal read pefore any thing hapend
    while np_all_acelerometro_terra[i][6] < start :
        y1.append(list(acelerometro_terra[i]))
        i = i + 1
    #get the event read
    while np_all_acelerometro_terra[i][6] >=  start and np_all_acelerometro_terra[i][6] <= end :
        y2.append(list(acelerometro_terra[i]))
        i = i + 1
    out_acelerometro_terra.append([y1,'normal'])
    out_acelerometro_terra.append([y2,lap])
    
#get the remander noemal read after any thing hapend
n=len(acelerometro_terra)
while i < n :
        y1.append(list(acelerometro_terra[i]))
        i = i + 1   
out_acelerometro_terra.append([y1,'normal'])
```
